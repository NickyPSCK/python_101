## Convert List to Pandas series

```python
date_list = ['20150101 00:00:00',
             '20150101 01:00:00',
             '20150101 02:00:00',
             '20150101 03:00:00',
             '20150101 04:00:00',
             '20150101 05:00:00',
             '20150101 06:00:00']
print(date_list)
```

```python
temp_list = [23, 25, 28, 23, 23, 25, 21]
print(temp_list)
```

```python
date_time = pd.to_datetime(date_list, format = '%Y-%m-%d %H:%M:%S')
print(date_time)
print(type(date_time))
```

```python
time_series = pd.Series(temp_list, index = date_time)
print(time_series.head())
type(time_series)
```

## Reindexing the Time Series Index

```python
date_1 = ['20150101 00:00:00',
             '20150101 01:00:00',
             '20150101 02:00:00',
             '20150101 03:00:00',
             '20150101 04:00:00',
             '20150101 05:00:00',
             '20150101 06:00:00']
temp_1 = [23, 25, 28, 23, 23, 25, 21]
dt_1 = pd.to_datetime(date_1, format = '%Y-%m-%d %H:%M:%S')
ts_1 = pd.Series(temp_1, index = dt_1)
print(ts_1)
```

```python
date_2 = date_1 = ['20150101 00:30:00',
             '20150101 01:00:00',
             '20150101 01:30:00',
             '20150101 02:00:00',
             '20150101 02:30:00',
             '20150101 03:00:00',
             '20150101 04:30:00',
             '20150101 05:00:00',
             '20150101 05:30:00',
             '20150101 06:00:00']
temp_2 = [23, 25, 28, 23, 23, 25, 21, 22, 24, 22]
dt_2 = pd.to_datetime(date_2, format = '%Y-%m-%d %H:%M:%S')
ts_2 = pd.Series(temp_2, index = dt_2)
print(ts_2)
```

```python
ts_3 = ts_1.reindex(ts_2.index)
print(ts_3)
```

```python
ts_4 = ts_1.reindex(ts_2.index, method='ffill')
print(ts_4)
```

```python
ts_5 = ts_1.reindex(ts_2.index, method='bfill')
print(ts_5)
```

```python
ts_sum1 = ts_1 + ts_2
print(ts_sum1)
```

```python
ts_sum2 = ts_4 + ts_1
print(ts_sum2)
```

## Resampling

```python
date_1 = ['20150101 00:00:00',
             '20150101 01:00:00',
             '20150101 02:00:00',
             '20150101 03:00:00',
             '20150101 04:00:00',
             '20150101 05:00:00',
             '20150101 06:00:00']
temp_1 = [23, 25, 28, 23, 23, 25, 21]
dt_1 = pd.to_datetime(date_1, format = '%Y-%m-%d %H:%M:%S')
ts_1 = pd.Series(temp_1, index = dt_1)
print(ts_1)
```

```python
ts_1.resample('D').sum()
```

```python
ts_1.resample('2H').mean()
```

```python
ts_1.resample('15min').first()
```

```python
ts_1.resample('15min').ffill()
```

```python
ts_1.resample('15min').first().interpolate('linear')
```
